```python
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

from pylab import rcParams
rcParams['figure.figsize'] = 16,9

```

```python
# Importing the dataset
data = pd.read_csv('xclara.csv')
print(data.shape)
data.head()
```

```python
# Getting the values and plotting it
f1 = data['V1'].values 
f2 = data['V2'].values 
X = np.array(list(zip(f1, f2)))
            # X , y co-ordinate
plt.scatter(f1, f2, c='black', s=7)
```

Now let's calucate : https://www.wikihow.com/Calculate-the-Sum-of-Squares-for-Error-(SSE)

```python
def SSE(Sx,Sy):
    Sx = sum(f1) / len(f1)
    Sy = sum(f2) / len(f2)
    Sx = f1 - Sx
    Sy = f2 - Sy
    sx = []
    sy = []
    for x , y in zip(Sx,Sy):
        sx.append(x*x)
        sy.append(y*y)
```

Now let's calucate EuclideanDistance : <br>
Source : https://people.revoledu.com/kardi/tutorial/Similarity/EuclideanDistance.html <br>
       : https://stackoverflow.com/questions/1401712/how-can-the-euclidean-distance-be-calculated-with-numpy/21986532

```python
# Euclidean Distance Caculator
def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)
```

```python
# Number of clusters
k = 3
# X coordinates of random centroids K times
C_x = np.random.randint(0, np.max(X)-20, size=k)
# Y coordinates of random centroids
C_y = np.random.randint(0, np.max(X)-20, size=k)
C = np.array(list(zip(C_x, C_y)), dtype=np.float32)
print(C)
```

```python
# Plotting along with the Centroids
plt.scatter(f1, f2, c='#050505', s=7)
plt.scatter(C_x, C_y, marker='*', s=200, c='g')
```

```python
# To store the value of centroids when it updates
C_old = np.zeros(C.shape)
C_old
```

```python
# Cluster Lables(0, 1, 2)
clusters = np.zeros(len(X))
clusters
```

```python
# Calling Euclidean Distance Caculator Error Function . - Distance between new centroids and old centroids
error = dist(C, C_old, None)
error
```

```python
# Loop will run till the error becomes zero
while error != 0:
    # Assigning each value to its closest cluster
    for i in range(len(X)):
        distances = dist(X[i], C) #  X[i] contains f1,f2 and C random init cordinates
        cluster = np.argmin(distances) #  get the minimum distance of all first occurence 
        clusters[i] = cluster
    # Storing the old centroid values
    C_old = deepcopy(C)
    # Finding the new centroids by taking the average value
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C[i] = np.mean(points, axis=0)
    error = dist(C, C_old, None)
    
```

```python
error
```

```python
colors = ['r', 'g', 'b', 'y', 'c', 'm']
fig, ax = plt.subplots()
for i in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
ax.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#000000')
```
